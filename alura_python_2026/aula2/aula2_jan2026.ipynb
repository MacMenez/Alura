```python
# Dicionário de renomeação
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Aplicando renomeação
df.rename(columns=novos_nomes, inplace=True)
df.columns

# Verificando resultado
#df.head()
```

```python
df["senioridade"].value_counts()
# BASE DO KAGGLE
# SE - SÊNIOR
# MI - MID (INTERMEDIÁRIO/PLENO)
# EN - ENTRY (INICIANTE/JUNIOR)
# EX - EXECUTIVO
```

```python
df["contrato"].value_counts()
# FT - FULL-TIME (INTEGRAL)
# CT - CONTRACT TEMPORARY (CONTRATO TEMPORÁRIO)
# PT - PART-TIME (MEIO-PERIODO)
# FL - FREELANCER
```

```python
df["remoto"].value_counts()
# 0 - presenciais
# 100 - remoto
# 50 - híbrido
```

```python
df["tamanho_empresa"].value_counts()
# M - midium (médio)
# L - large (grande)
# S - small (pequeno)
```

```python
senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M':	'media'

}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
mapa_trabalho = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}

df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['remoto'].value_counts()
```

```python
import numpy as np # Biblioteca para manipulaçao numérica

# criando um df - dataframe do zero -> dataframe de teste
df_salarios = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# Calcula média salarial e substitui os nulos pela média e arredonda
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2)) 

''''
fillna -> fill = preencher -> preencher os valores nulos
mean -> média
round -> arredondamento
'''
```

```python
# Exibir tabela
df_salarios
```

```python
# Calculo com Mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median()) 
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Qurta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill() # ffill -> Foward Fill = completar com o valor anterior

df_temperaturas # Exibir
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terça", "Qurta", "Quinta", "Sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill() # bfill -> Back Fill = completar com o valor posterior

df_temperaturas # Exibir
```

```python
# Preencher com valor fixo
df_cidades = pd.DataFrame({
    'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
    'cidade': ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})
df_cidades['cidade_preenchida'] = df_cidades["cidade"].fillna("Não Informado")
display(df_cidades)

'''
Métodos usa-se parênteses -> ()
Criando um DataFrame/Base de Dados usa-se chaves -> {}
Coluna específica da base de dados usa-se colchetes -> []
'''
```
